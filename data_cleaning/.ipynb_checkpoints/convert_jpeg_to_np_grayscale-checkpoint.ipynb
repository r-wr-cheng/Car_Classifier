```python
from PIL import Image
import numpy as np

def load_image(infilename) :
    img = Image.open(infilename)
    img.load()
    #convert to grayscale
    data = np.asarray(img.convert('L'))
    return data
```

```python
grayscale_dict = {}

path_pre = '/home/ec2-user/raw_data/cars_train/'
for i in os.listdir(path_pre):
    print(i)
    grayscale_dict[i] = load_image(path_pre + i)
```

```python
p.dump(grayscale_dict, open('/home/ec2-user/grayscale_dict.p', 'wb'))
```
