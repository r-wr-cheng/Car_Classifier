```python
from PIL import Image
import numpy as np

def load_image(infilename) :
    img = Image.open(infilename)
    img.load()
    #convert to grayscale
    data = np.asarray(img.convert('L'))
    return data
```

```python
load_image('/home/ec2-user/00001.jpg')
```
